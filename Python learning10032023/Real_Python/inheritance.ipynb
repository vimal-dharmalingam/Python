```python
class person:
    
    description='Person'
    
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    def speak(self):
        print(self.name,self.age)
        
    def eat(self,food):
        
        print(self.name, food)
        
    def action(self):
        
        print(self.name, self.age)
        
class baby(person):
    
    description=' baby'
    def speak(self):
        
        print('ba ba ba ba')
        
    def nap(self):
        
        print(self.name)
        
        
        
        
        
```

```python
persn=person('vimal',20)
persn.speak()
```

```python
persn.eat('rice')
```

```python
persn.action()
```

```python
bab=baby('vicky',25)
bab.nap()
```

```python
bab.speak()
```

```python
bab.eat('newrice')
```

```python
persn.description
```

```python
bab.description
```

```python
isinstance(bab,person) # the bab object is the class person
```

```python
class Family:
    def __init__(self, lastname, member_count, hair_color):
        self.lastname = lastname
        self.member_count = member_count
        self.hair_color = hair_color
        

class Person(Family):
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age
```

```python
family=Family('vimal',10, 'black')
family.lastname
```
