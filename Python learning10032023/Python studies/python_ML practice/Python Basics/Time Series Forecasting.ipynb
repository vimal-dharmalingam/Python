#ARIMA Model - AR, I and MA

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
#p - periods taken for autoregressive, d = period lag/difference and q = periods in moving average model
mode_arima = ARIMA(train, order = (0,1,3)) #(p,d,q)
mode_arima_fit = mode_arima.fit()
```

```python
predictions = mode_arima_fit.forecast(steps=12)[0]
```

```python
import itertools
p=q=d=range(0,5)
pdq = list(itertools.product(p,d,q))
pdq
```

```python
import warnings 
warnings.filterwarnings('ignore')
for param in pdq:
    try:
        mode_arima = ARIMA(train, order = param) #(p,d,q)
        mode_arima_fit = mode_arima.fit()
        print(param, mode_arima_fit.aic)
    except:
        continue
```
