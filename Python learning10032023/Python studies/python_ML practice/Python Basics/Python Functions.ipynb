```python
def Headcount(calls, AHT, Occupancy, Schedule_Efficiency, Shrinkage, Workdays, LoginHours):
    return (calls*AHT/3600/Occupancy/Schedule_Efficiency/(1-Shrinkage))/(Workdays*LoginHours)
```

```python
Headcount(150000,300,0.75,0.90,0.20,20,7.5)
```
