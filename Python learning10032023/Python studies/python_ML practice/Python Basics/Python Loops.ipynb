```python
#For loop :For loops are called iterators, it iterates the element based on the condition set
for x in range(1 , 101):
    print(x)
```

```python
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
for m in month:
    print(m)
```

```python
#Break Statement in For loop
for x in range(1 , 101):
    if (x==80):break
    if(x%2<>0):continue   #Even number 
    print(x)
```

```python
for x in range(1 , 101):
    if (x==80):break
    if(x%2==0):continue   #odd number 
    print(x)
```

```python
#enumerate function in for loop -1)It returns the index number for the member 2)And the member of the collection that we are looking at
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
for i, m in enumerate(month):
    print(i,m)
```
