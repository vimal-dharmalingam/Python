```python
import pandas as pd
import os

loc = r'C:\Users\ManagingDirector\Documents\ML Implementation\NLP\BBC News\bbc-fulltext\business'
os.chdir(loc)
filelist = os.listdir()

data = []
path = loc
files = [f for f in os.listdir(path) if os.path.isfile(f)]
for f in files:
    with open(f,'r') as myfile:
        data.append(myfile.read())

business = pd.DataFrame(data)
print (business.shape)
#Tagging under Business category
business['Label'] = 'Business'
business
```
